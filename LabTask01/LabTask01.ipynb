### Part 7: Mapping and Transformation ###
1. Convert categorical values in the gender column to numeric codes.

2. Create a new column that categorizes age into groups (e.g., child, adult, senior).



```python
df_recode= df.copy()
df_recode["Sex"].replace(['female', 'male'], [0, 1], inplace=True)
df_recode
```

```python
def age_category(age): # function that takes age as var and return a category depending on the value 
    if age < 18: #if age less than 18 category is a child
        return "child"
    elif 18 <= age < 70: #if age more than 18 and less than 70 category is a adult
        return "adult"
    else: # more than 70 category is senior
        return "senior"
df["age_category"] = df["Age"].map(age_category) #map is used here to apply the function to every value in the age column
df
```
